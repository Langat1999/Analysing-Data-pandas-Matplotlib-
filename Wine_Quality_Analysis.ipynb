# Wine Quality Data Analysis

This notebook demonstrates how to load, explore, analyze, and visualize the wine quality datasets using pandas, matplotlib, and seaborn. The workflow includes data loading, cleaning, basic statistics, group analysis, and multiple types of visualizations with customizations and findings.

## 1. Import Required Libraries
Import pandas, numpy, matplotlib, seaborn, and any other necessary libraries for data analysis and visualization.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## 2. Load the Dataset from Local Directory
We will load the red wine quality dataset. Error handling is included for file reading.

```python
# Load the red wine quality dataset
try:
    df = pd.read_csv('winequality-red.csv', sep=';')
    print("Red wine dataset loaded successfully.")
except FileNotFoundError:
    print("File not found. Please check the file path.")
    df = None
except Exception as e:
    print(f"Error loading dataset: {e}")
    df = None
```

## 3. Display First Few Rows
Display the first five rows of the dataset to inspect the data.

```python
# Display the first five rows
if df is not None:
    display(df.head())
else:
    print("DataFrame is empty.")
```

## 4. Explore Dataset Structure
Check the data types of each column and identify missing values.

```python
# Check data types and missing values
if df is not None:
    display(df.info())
    print("\nMissing values per column:")
    print(df.isnull().sum())
else:
    print("DataFrame is empty.")
```

## 2b. Load the White Wine Dataset
We will also load the white wine quality dataset for comparison.

```python
# Load the white wine quality dataset
try:
    df_white = pd.read_csv('winequality-white.csv', sep=';')
    print("White wine dataset loaded successfully.")
except FileNotFoundError:
    print("File not found. Please check the file path.")
    df_white = None
except Exception as e:
    print(f"Error loading dataset: {e}")
    df_white = None
```

## 3b. Display First Few Rows of White Wine Data
Display the first five rows of the white wine dataset to inspect the data.

```python
# Display the first five rows of white wine data
if df_white is not None:
    display(df_white.head())
else:
    print("White wine DataFrame is empty.")
```

## 4b. Explore White Wine Dataset Structure
Check the data types of each column and identify missing values in the white wine dataset.

```python
# Check data types and missing values for white wine
def explore_structure(df, name):
    if df is not None:
        print(f"\n{name} info:")
        display(df.info())
        print("\nMissing values per column:")
        print(df.isnull().sum())
    else:
        print(f"{name} DataFrame is empty.")

explore_structure(df_white, "White wine")
```

## 5. Handle Missing Values
Clean both datasets by dropping rows with missing values (if any).

```python
# Drop rows with missing values in both datasets
df_clean = df.dropna() if df is not None else None
df_white_clean = df_white.dropna() if df_white is not None else None

print(f"Red wine: {len(df) - len(df_clean)} rows dropped.")
print(f"White wine: {len(df_white) - len(df_white_clean)} rows dropped.")
```

## 6. Compute Basic Statistics
Compute mean, median, standard deviation, and other statistics for numerical columns in both datasets.

```python
# Basic statistics for red wine
def show_stats(df, name):
    if df is not None:
        print(f"\n{name} - Describe:")
        display(df.describe())
        print(f"\n{name} - Median:")
        print(df.median(numeric_only=True))
    else:
        print(f"{name} DataFrame is empty.")

show_stats(df_clean, "Red wine")
show_stats(df_white_clean, "White wine")
```

## 7. Group and Aggregate Data
Group by the 'quality' column and compute the mean of numerical columns for each group in both datasets.

```python
# Group by 'quality' and compute mean for each group
def group_by_quality(df, name):
    if df is not None:
        grouped = df.groupby('quality').mean(numeric_only=True)
        print(f"\n{name} - Mean values by quality:")
        display(grouped)
        return grouped
    else:
        print(f"{name} DataFrame is empty.")
        return None

grouped_red = group_by_quality(df_clean, "Red wine")
grouped_white = group_by_quality(df_white_clean, "White wine")
```

## 8. Identify Patterns and Insights
Summarize interesting findings from the statistics and group analysis.

- Both datasets show a range of quality scores, with most wines clustered around the middle values.
- Mean values of features such as alcohol, sulphates, and volatile acidity vary with quality, indicating their influence on wine quality.
- White wines generally have higher residual sugar and lower volatile acidity compared to red wines.
- Median and mean values are close for most features, suggesting relatively symmetric distributions.

## 9. Line Chart: Trends Over Quality
Visualize how alcohol content changes with wine quality for both datasets.

```python
# Line chart: Alcohol content vs. quality
plt.figure(figsize=(10,5))
plt.plot(grouped_red.index, grouped_red['alcohol'], marker='o', label='Red wine')
plt.plot(grouped_white.index, grouped_white['alcohol'], marker='o', label='White wine')
plt.title('Alcohol Content vs. Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Mean Alcohol Content')
plt.legend()
plt.show()
```

## 10. Bar Chart: Average Sulphates by Quality
Compare the average sulphates content for each quality score in both datasets.

```python
# Bar chart: Average sulphates by quality
# Align quality indices for both datasets
all_qualities = sorted(set(grouped_red.index).union(set(grouped_white.index)))
red_sulphates = grouped_red['sulphates'].reindex(all_qualities, fill_value=0)
white_sulphates = grouped_white['sulphates'].reindex(all_qualities, fill_value=0)

bar_width = 0.35
index = np.arange(len(all_qualities))

plt.figure(figsize=(10,5))
plt.bar(index, red_sulphates, bar_width, label='Red wine')
plt.bar(index + bar_width, white_sulphates, bar_width, label='White wine')
plt.xlabel('Quality')
plt.ylabel('Average Sulphates')
plt.title('Average Sulphates by Wine Quality')
plt.xticks(index + bar_width/2, all_qualities)
plt.legend()
plt.show()
```

## 11. Histogram: Distribution of Alcohol Content
Visualize the distribution of alcohol content for both red and white wines.

```python
# Histogram: Distribution of alcohol content
plt.figure(figsize=(10,5))
sns.histplot(df_clean['alcohol'], bins=20, color='red', label='Red wine', kde=True, stat='density', alpha=0.6)
sns.histplot(df_white_clean['alcohol'], bins=20, color='gold', label='White wine', kde=True, stat='density', alpha=0.6)
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## 12. Scatter Plot: Relationship Between Alcohol and Quality
Visualize the relationship between alcohol content and quality for both datasets.

```python
# Scatter plot: Alcohol vs. Quality
plt.figure(figsize=(10,5))
sns.scatterplot(x='alcohol', y='quality', data=df_clean, color='red', label='Red wine', alpha=0.5)
sns.scatterplot(x='alcohol', y='quality', data=df_white_clean, color='gold', label='White wine', alpha=0.5)
plt.title('Alcohol Content vs. Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.legend()
plt.show()
```
